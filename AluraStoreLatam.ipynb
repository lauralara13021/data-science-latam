### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#ingreso total tienda
precio_tienda = float(tienda['Precio'].sum())
precio_tienda2 = float(tienda2['Precio'].sum())
precio_tienda3 = float(tienda3['Precio'].sum())
precio_tienda4 = float(tienda4['Precio'].sum())

variables = {
 "tienda1": precio_tienda,
 "tienda2": precio_tienda2,
 "tienda3": precio_tienda3,
 "tienda4": precio_tienda4
}

for nombre, valor in variables.items():
  print(f"{nombre}: {valor}")

print(max(variables, key=variables.get))
print(min(variables, key=variables.get))
```

El objetivo de este análisis era analizar la tienda que se pudiera vender entre 4 tiendas pertenecientes al señor Juan.
- La tienda con mayores ingresos es la tienda 1 y la tienda con menores ingresos es la tienda 4.
- La tienda con satisfacción mayor es la tienda 2 y la tienda con menor satisfacción es la tienda 1.
- La tienda con mayor costo de envío promedio es la tienda 1 y con menor costo de envío promedio es la tienda 3

No hay una tienda que lidere en varios aspectos, por ejemplo, una tienda puede tener mayores ingresos pero un bajo índice de satisfacción de clientes. Por ello se propone un índice que cosidera diferentes pesos a los parámetros estipulados.

Indice = 0.4*ingresos + 0.3*satisfaction - 0.3*costo_envio

```python
variables_norm = {}
min_val = min(variables.values())
max_val = max(variables.values())
for nombre, valor in variables.items():
    norm = (valor - min_val) / (max_val - min_val)
    variables_norm[nombre] = norm

for nombre, valor in variables_norm.items():
    print(f"{nombre}: {valor:.3f}")
```

```python
satisfactions_norm = {}

min_val = min(satisfactions.values())
max_val = max(satisfactions.values())

for nombre, valor in satisfactions.items():
    norm = (valor - min_val) / (max_val - min_val)
    satisfactions_norm[nombre] = norm

for nombre, valor in satisfactions_norm.items():
    print(f"{nombre}: {valor:.3f}")
```

```python
promedios_norm = {}
min_val = min(promedios.values())
max_val = max(promedios.values())
for nombre, valor in promedios.items():
    norm = (valor - min_val) / (max_val - min_val)
    promedios_norm[nombre] = norm

for nombre, valor in promedios_norm.items():
    print(f"{nombre}: {valor:.3f}")
```

```python

indice = {}

for nombre_tienda in variables_norm.keys():

    ingresos = variables_norm[nombre_tienda]
    satisfaccion = satisfactions_norm[nombre_tienda]
    envio = promedios_norm[nombre_tienda]

    envio_score = 1 - envio

    score = 0.4*ingresos + 0.3*satisfaccion + 0.3*envio_score

    indice[nombre_tienda] = score

for nombre, valor in indice.items():
    print(f"{nombre}: {valor:.3f}")

print("Mejor tienda:", max(indice, key=indice.get))
print("Peor tienda:", min(indice, key=indice.get))
```

La tienda que debería vender el sr. Juan para poder comprar su emprendimiento es la tienda 4, pues el índice que evalua que tan buena es la tienda, es inferior a todas las otras tiendas en la tienda 4.
